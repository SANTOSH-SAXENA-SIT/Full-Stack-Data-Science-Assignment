## Questions

1. Write a Python program to convert kilometers to miles?
2. Write a Python program to convert Celsius to Fahrenheit?
3. Write a Python program to display calendar?
4. Write a Python program to solve quadratic equation?
5. Write a Python program to swap two variables without temp variable?

```python
# Module declaration#
#-------------------#
import calendar
#-------------------#

class Assignment_2:
    """
    This module is supposed to solve the above mentioned task 
    
    Developed by : Santosh Saxena
    Date : 2/10/2020
    """
    
    def question_1(self, kilometer):
        """
        This module will convert kilometers to miles
        
        Argument list 
        
        kilometer : Value of kilometer decided by the user at the run time
        """
        try :
            return 0.621371*kilometer
        except Exception as e :
            print("This code failed because of : ", str(e))
            return -1
    
    def question_2(self, celcius):
        """
        This module will convert celsius to fahrenheit
        
        Argument list
        
        celcius : value of celcius decided by the user at the run time
        """
        try:
            return (celcius * (9/5)) + 32
        except Exception as e:
            print("This code failed because of : ", str(e))
            return -1
    
    def question_3(self, year , month):
        """
        This module will show the calendar accordin to the year and month.
        
        Argument list
        
        year : Value of year given by the user at the runtime. 
        month : Value of month given by the user at the runtime.

        """
        try:
            return(calendar.month(year, month))
        except Exception as e:
            print("The code failed because of : ", str(e))
            return -1
    
    def question_4(self,a,b,c):
        """
        This module will solve the quadratic equation 
        
        Argument list
        
        a,b,c (ax^2 + bx + c) : The value of a,b,c will be entered by the user at the runtime. 
        
        """
        try:
            d = (b**2) - 4*a*c
            return ((-b + d**(0.5))/2*a , (-b - d**(0.5))/2*a)
        except Exception as e:
            print("The code failed because of : ", str(e))
            return -1
        
    def question_5(self, first_variable , second_variable):
        """
        This module will swap variables
        
        Argument list
        
        first variable : first value given by the user
        second variable : second value given by the user
        
        """
        try:
            first_variable = first_variable + second_variable
            second_variable = first_variable - second_variable
            first_variable = first_variable - second_variable
            return (first_variable, second_variable)
        except Exception as e:
            print("The code failed because of ", str(e))
            return -1
                
```

```python
# Testing Case 1

obj = Assignment_2()
print("Question 1 : ",obj.question_1(10))
print("Question 2 : ",obj.question_2(10))
print("Question 3 : ",obj.question_3(2000,10))
print("Question 4 : ",obj.question_4(1,2,3))
print("Question 5 : ",obj.question_5(10,20))
```
